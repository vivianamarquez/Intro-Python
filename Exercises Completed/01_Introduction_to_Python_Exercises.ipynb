## 1.1 — Podcast Episode

You are tracking a podcast episode.

Create variables for:
- `episode_number` (int)
- `duration_minutes` (float)
- `is_finished` (bool)
- `show_title` (str)

Print one sentence that combines all four, for example:
`You finished episode 12 of Science Weekly (38.5 min).`

Use an f-string and change the wording if `is_finished` is `False`.

````python
episode_number = 12
duration_minutes = 38.5
is_finished = True
show_title = "Science Weekly"

if is_finished:
    print(f"You finished episode {episode_number} of {show_title} ({duration_minutes} min).")
else:
    print(f"You started episode {episode_number} of {show_title} ({duration_minutes} min).")
````

## 1.2 — Celsius to Fahrenheit

The formula is:

`fahrenheit = (celsius * 9/5) + 32`

Store `celsius = 22` and calculate the equivalent Fahrenheit value. Print the result rounded to one decimal place using `round(value, 1)`.

````python
celsius = 22
fahrenheit = (celsius * 9 / 5) + 32
print(round(fahrenheit, 1))
````

## 1.3 — Splitting the Bill

Four friends went out for dinner. The total bill is `$87`.

1. Store the bill and number of people in variables.
2. Use floor division (`//`) to find how many whole dollars each person pays.
3. Use modulo (`%`) to find how many dollars are left over.
4. Print all three results with short labels.

````python
bill = 87
people = 4
per_person = bill // people
leftover = bill % people

print(f"Bill: ${bill}")
print(f"Each person pays: ${per_person}")
print(f"Leftover: ${leftover}")
````

## 1.6 — Password Check

A password is considered acceptable when:
- it has at least 8 characters, **and**
- it is not the word `"password"`

Test these two passwords and print `True` or `False` for each:
- `"sunset42"`
- `"password"`

Hint: `len()` counts characters.

````python
passwords = ["sunset42", "password"]

for password in passwords:
    acceptable = len(password) >= 8 and password != "password"
    print(acceptable)
````
