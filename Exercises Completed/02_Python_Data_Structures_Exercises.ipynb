## 2.1 — Playlist Slicing

A playlist is stored as one long string:

```python
playlist = "Neon Sky|Midnight Drive|Golden Hour|Static Bloom|Paper Moon"
```

Song titles are separated by `|`.

1. Use indexing to print the first song title (hint: slice up to the first `|`).
2. Use slicing to print the last song title.
3. Print how many characters are in the full playlist string.

````python
playlist = "Neon Sky|Midnight Drive|Golden Hour|Static Bloom|Paper Moon"

first_pipe = playlist.index("|")
print(playlist[:first_pipe])

last_pipe = playlist.rfind("|")
print(playlist[last_pipe + 1:])

print(len(playlist))
````

## 2.2 — Clean Up a Username

A user registered with messy input:

```python
username = "  @NovaCoder  "
```

1. Remove extra spaces.
2. Remove the `@` symbol.
3. Convert the result to lowercase.
4. Print the final username.

````python
username = "  @NovaCoder  "

username = username.strip()
username = username.replace("@", "")
username = username.lower()

print(username)
````

## 2.3 — Top Three Scores

A game stores high scores in a list:

```python
scores = [420, 315, 510, 280, 505, 190]
```

1. Print the highest score (use `max()`).
2. Print the lowest score (use `min()`).
3. Add a new score of `390` to the list.
4. Print the top 3 scores using slicing and `sorted()`.

````python
scores = [420, 315, 510, 280, 505, 190]

print(max(scores))
print(min(scores))

scores.append(390)
print(sorted(scores, reverse=True)[:3])
````

## 2.4 — Record a Match Result

A soccer match result should be stored as a tuple so it cannot be changed later:

```python
match = ("Tigers", "Eagles", 2, 1)
```

Unpack the tuple into four variables: `home_team`, `away_team`, `home_goals`, `away_goals`.

Print a one-line summary like:
`Tigers 2 - 1 Eagles`

````python
match = ("Tigers", "Eagles", 2, 1)
home_team, away_team, home_goals, away_goals = match

print(f"{home_team} {home_goals} - {away_goals} {away_team}")
````

## 2.5 — Overlapping Skills

Two project teams listed their skills:

```python
team_a = ["python", "git", "sql", "python", "figma"]
team_b = ["sql", "docker", "python", "writing"]
```

1. Turn each list into a set.
2. Find skills both teams share.
3. Find every unique skill across both teams.
4. Find skills only `team_a` has (hint: set difference with `-`).

````python
team_a = ["python", "git", "sql", "python", "figma"]
team_b = ["sql", "docker", "python", "writing"]

skills_a = set(team_a)
skills_b = set(team_b)

print(skills_a & skills_b)
print(skills_a | skills_b)
print(skills_a - skills_b)
````

## 2.6 — Contact Card

Build a dictionary called `contact` for one person with keys:
- `name`
- `email`
- `city`
- `phone`

Fill in your own sample values. Then:
1. Print only the email.
2. Update the city to a new value.
3. Loop through the dictionary and print each key and value on its own line.

````python
contact = {
    "name": "Maria Lopez",
    "email": "maria@example.com",
    "city": "Bogota",
    "phone": "555-0123",
}

print(contact["email"])
contact["city"] = "Cali"

for key, value in contact.items():
    print(f"{key}: {value}")
````

## 2.7 — Build a Playlist from a Dictionary

Song lengths are stored like this:

```python
song_lengths = {
    "Neon Sky": 204,
    "Midnight Drive": 187,
    "Golden Hour": 231,
}
```

Values are lengths in seconds.

1. Print the length of `"Golden Hour"`.
2. Add a new song `"Static Bloom"` with length `198`.
3. Use a loop to print each song and its length in the format: `Golden Hour: 231 sec`.

````python
song_lengths = {
    "Neon Sky": 204,
    "Midnight Drive": 187,
    "Golden Hour": 231,
}

print(song_lengths["Golden Hour"])
song_lengths["Static Bloom"] = 198

for song, length in song_lengths.items():
    print(f"{song}: {length} sec")
````
