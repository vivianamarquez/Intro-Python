## 4.1 — Tip Calculator

Write a function `calculate_tip(bill, percent)` that returns the tip amount.

- `bill` is the total in dollars
- `percent` is the tip percentage (e.g. `18` for 18%)

Test with a `$54` bill and a `20%` tip. Print the tip rounded to 2 decimal places.

````python
def calculate_tip(bill, percent):
    return bill * percent / 100

tip = calculate_tip(54, 20)
print(round(tip, 2))
````

## 4.2 — Temperature Label

Write a function `temperature_label(celsius)` that **returns** a string:
- `"freezing"` if celsius is at or below 0
- `"cool"` if above 0 and at most 15
- `"warm"` if above 15 and at most 28
- `"hot"` if above 28

Test with `-3`, `12`, and `31`.

````python
def temperature_label(celsius):
    if celsius <= 0:
        return "freezing"
    if celsius <= 15:
        return "cool"
    if celsius <= 28:
        return "warm"
    return "hot"

for temp in [-3, 12, 31]:
    print(temp, temperature_label(temp))
````

## 4.3 — Count Vowels

Write a function `count_vowels(text)` that returns how many vowels (`a`, `e`, `i`, `o`, `u`) appear in `text`. Count both uppercase and lowercase.

Test with `"Data Science"`.

````python
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count

print(count_vowels("Data Science"))
````

## 4.4 — Roll Two Dice

Use the `random` module to simulate rolling two six-sided dice 5 times.

Each roll should print something like:
`Roll 1: 4 + 2 = 6`

At the end, print how many rolls had a total of 7.

````python
import random

sevens = 0

for roll_number in range(1, 6):
    die_1 = random.randint(1, 6)
    die_2 = random.randint(1, 6)
    total = die_1 + die_2
    print(f"Roll {roll_number}: {die_1} + {die_2} = {total}")
    if total == 7:
        sevens += 1

print(f"Rolls with total 7: {sevens}")
````

## 4.5 — Days Until Deadline

Use `datetime` to:

1. Create a deadline: `December 15, 2026`
2. Get today's date
3. Calculate and print how many days remain until the deadline

If the deadline has already passed, print `Deadline passed` instead.

````python
import datetime

deadline = datetime.datetime(2026, 12, 15)
today = datetime.datetime.today()
days_remaining = (deadline - today).days

if days_remaining < 0:
    print("Deadline passed")
else:
    print(days_remaining)
````

## 4.6 — Test Score Stats

Import `numpy` as `np`.

## 4.7 — Receipt Formatter

Write a function `format_receipt(items)` where `items` is a dictionary mapping item names to prices.

The function should return a single string with one item per line, like:

```
Coffee: $4.50
Muffin: $3.25
Total: $7.75
```

Test with `{"Coffee": 4.50, "Muffin": 3.25}`.

````python
def format_receipt(items):
    lines = []
    total = 0
    for name, price in items.items():
        lines.append(f"{name}: ${price:.2f}")
        total += price
    lines.append(f"Total: ${total:.2f}")
    return "\n".join(lines)

print(format_receipt({"Coffee": 4.50, "Muffin": 3.25}))
````
