## 3.1 — What to Wear

Set `temperature = 18` and `is_raining = True`.

Write conditionals that print clothing advice:
- If it is raining → print `Bring an umbrella`
- Else if temperature is below 10 → print `Wear a heavy coat`
- Else if temperature is below 20 → print `Wear a light jacket`
- Else → print `T-shirt weather`

More than one message can print.

````python
temperature = 18
is_raining = True

if is_raining:
    print("Bring an umbrella")

if temperature < 10:
    print("Wear a heavy coat")
elif temperature < 20:
    print("Wear a light jacket")
else:
    print("T-shirt weather")
````

## 3.2 — Letter Grade

Write code that converts a numeric `grade` into a letter:

| Score | Letter |
|-------|--------|
| 90+   | A      |
| 80-89 | B      |
| 70-79 | C      |
| 60-69 | D      |
| below 60 | F   |

Test with `grade = 84` and print the letter.

````python
grade = 84

if grade >= 90:
    letter = "A"
elif grade >= 80:
    letter = "B"
elif grade >= 70:
    letter = "C"
elif grade >= 60:
    letter = "D"
else:
    letter = "F"

print(letter)
````

## 3.3 — Running Total

Start with this list of daily step counts:

```python
steps = [3200, 5100, 4800, 8900, 2500]
```

Use a `for` loop to:
1. Print each day's steps with its day number (`Day 1: 3200`, etc.).
2. Keep a running total and print the total at the end.

````python
steps = [3200, 5100, 4800, 8900, 2500]
total = 0

for day_number in range(1, len(steps) + 1):
    daily_steps = steps[day_number - 1]
    print(f"Day {day_number}: {daily_steps}")
    total += daily_steps

print(f"Total: {total}")
````

## 3.4 — Multiplication Table

Pick a number `n = 7`.

Use a `for` loop with `range()` to print the multiplication table from 1 to 10:

```
7 x 1 = 7
7 x 2 = 14
...
7 x 10 = 70
```

````python
n = 7

for multiplier in range(1, 11):
    print(f"{n} x {multiplier} = {n * multiplier}")
````

## 3.5 — Guessing Game Setup

Simulate someone guessing a PIN. The correct PIN is `4`.

```python
attempts = [1, 8, 4]
```

Use a `while` loop to go through the attempts one at a time:
- Print `Try again` for wrong guesses
- Print `Access granted` and stop when the correct PIN is found

Do not write an infinite loop.

````python
correct_pin = 4
attempts = [1, 8, 4]
index = 0

while index < len(attempts):
    guess = attempts[index]
    if guess == correct_pin:
        print("Access granted")
        break
    else:
        print("Try again")
    index += 1
````

## 3.6 — Filter Long Words

Given:

```python
words = ["ai", "python", "loop", "data", "go", "structure"]
```

1. Use a `for` loop to build a new list of words with more than 4 letters.
2. Rewrite the same logic as a **list comprehension**.
3. Print both results.

````python
words = ["ai", "python", "loop", "data", "go", "structure"]

long_words = []
for word in words:
    if len(word) > 4:
        long_words.append(word)

long_words_comp = [word for word in words if len(word) > 4]

print(long_words)
print(long_words_comp)
````

## 3.7 — Celsius to Fahrenheit (Loop Edition)

Convert every temperature in this list to Fahrenheit and store the results in a new list:

```python
celsius_temps = [0, 10, 20, 30, 40]
```

Use the formula `f = (c * 9/5) + 32`.

Try it first with a `for` loop, then again with a list comprehension. Print both result lists.

````python
celsius_temps = [0, 10, 20, 30, 40]

fahrenheit_loop = []
for temp in celsius_temps:
    fahrenheit_loop.append((temp * 9 / 5) + 32)

fahrenheit_comp = [(temp * 9 / 5) + 32 for temp in celsius_temps]

print(fahrenheit_loop)
print(fahrenheit_comp)
````
