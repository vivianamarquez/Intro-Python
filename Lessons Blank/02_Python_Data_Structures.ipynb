**Other functions related to text strings**

- `str()` - Converts an integer/float into a text string.
- `int()` - Converts a text string into an integer.
- `float()` - Converts a text string into a floating-point number.
- `len()` - Tells you how many characters are in a text string.
- `.upper()` - Converts all letters in a text string to UPPERCASE.
- `.lower()` - Converts all letters in a text string to lowercase.
- `.swapcase()` - Converts uppercase letters to lowercase and lowercase letters to uppercase.
- `.capitalize()` - Converts the first letter of a text string to uppercase.
- `.strip()` - Removes any whitespace from the beginning or end of a text string.
- `.split(",")` - Splits a text string into substrings.
- `.startswith()` - Returns true if the text string starts with the specified value.
- `.endswith()` - Returns true if the text string ends with the specified value.
- `.index()` - Finds the first occurrence of the specified value.
- `.isnumeric()` - Checks whether the text string contains numeric characters.
- `==` - Checks whether 2 text strings are equal.
- `!=` - Checks whether 2 text strings are NOT equal.
- `in` / `not in` - Checks whether a substring is or is not inside the text string.

## 2.1 Lists


A list keeps items in order. Each position has an index:

```python
values = [1, 3, 5, 2, 4]
# index:  0  1  2  3  4
```

- Mutable: it can change!
- Declare it inside square brackets `[` `]`.
- Each value in a list is called an element.

To find its size

To extract its elements

**What can go inside a list?**

Anything!
- Integers
- Decimals
- Text strings
- Booleans
- Other lists
- A mix of everything

## 2.2 Tuples



A tuple is like a list, but with two differences:

- It is not mutable, so it cannot be changed.
  - You cannot add new elements.
  - You cannot delete elements.
  - You cannot modify elements.
- It is declared with parentheses `(` `)`.

**Advantages**
- Tuples are faster than lists because they are not dynamic.
- They make your code safer for data that does not need to be modified.

## 2.3 Sets



- All elements are unique.
- Order does not matter.
- A set is declared with curly braces `{` `}`.

## 2.4 Dictionaries



A dictionary stores **key-value pairs**:

```python
info = {"first_name": "Viviana", "last_name": "Marquez"}
```

Keys are unique.
