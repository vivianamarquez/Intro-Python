# 1. ⚙️ Functions


Functions are pieces of code that you can use again and again instead of writing them multiple times. They make your life easier!


Ingredients of a function:
- Input data
- A name
- Instructions
- Result

A function has four main ingredients:

```python
def function_name(input_data):
    # instructions
    return result
```

- The function name tells Python which reusable block to run.
- The input data goes inside the parentheses.
- The indented code block contains the instructions.
- `return` sends the result back.

To use a function that already exists, call its name and pass values inside the parentheses:

```python
function_name("hello", 3141592, [23, 7])
```

````python
def say_hello(name):
  print(f"Hello {name}!")
  print("How are you?")
````

````python
say_hello("Victoria")
````

````python
def add_seven(number):
  return number + 7
````

````python
add_seven(10)
````

````python
favorite_number = 23
add_seven(favorite_number)
````

````python
hello_john = say_hello("John")
add_6_7 = add_seven(6)
````

````python
hello_john
````

````python
add_6_7
````

### Time

````python
import time

# Calculate the execution time of a code block.
start_time = time.time()

total = 0
for i in range(50000):
    total = total + i

# Elapsed time
elapsed_time = time.time() - start_time
print(elapsed_time)
````

````python
for student in students[:5]:
    print(student)
    time.sleep(2)
````
