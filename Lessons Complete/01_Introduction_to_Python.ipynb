# 2. Variables and Data Types in Python

- Variables are the storage place and name for a piece of data.
- You can change the value of a variable at any time.

### Basic Python Data Types:

- `int`: Whole numbers
Example: `23`

- `float`: Decimal numbers
Example: `3.14`

- `boolean`: True or false value (`True` | `False`)
Example: `True`

- `string`: A sequence of characters inside single or double quotes
Example: `'python'` or `"Hello World"`

```python
favorite_number = 23
```

### Booleans

- Booleans can only have two values: `True` or `False`.
- `True` is not the same as `true`. The capital first letter matters.

```python
5 == 5
```

### 🤔 What data type do we have?

You can check the data type with the built-in `type()` function.

```python
type(True)
```

### Review: What Is a Variable?

- You can think of variables as a name attached to a value.
- To create a variable in Python, we do not need to declare its type ahead of time.
- Variables help us organize all the data around us.
- You can name a variable almost anything you want, but it is a good idea to use short, descriptive names.
- A variable name cannot contain spaces or start with numbers. In Python, variable names are usually lowercase and use underscores.

# 3. Arithmetic Syntax

In Python, there are specific ways to write arithmetic formulas. Addition, subtraction, and multiplication work the way you probably expect.

```python
a = 20
b = 10

a1 = b + a
a2 = a - b
a3 = a * b

print(a1)
print(a2)
print(a3)
```

There are two types of division.

True division: use one slash `/`. The result is a decimal number.

```python
10 / 3
```

Floor division: use two slashes `//`. The result is an integer: the quotient of `a` and `b` with the fractional part removed.

```python
10 // 3
```

The percent sign `%` is called the **modulo** operator. It returns the remainder after dividing two numbers.

```python
10 % 3
```

To use exponents, or raise something to a power, use a double asterisk `**`.

```python
2 ** 3
```

# 4. Comparisons

Two equals signs `==` compare the values on both sides and decide the relationship between them. These are called comparison or relational operators.

Comparison operators return booleans. Booleans are `True` or `False` values.

```python
a = 10
b = 20

a == b
```

Not equal is represented with `!=`.

```python
10 != 20
```

Less-than `<` and greater-than `>` signs

```python
3 < 20
```

Less-than-or-equal-to `<=` and greater-than-or-equal-to `>=` signs

```python
3 >= 20
```

# 5. Logical Operators

From highest precedence to lowest: `NOT`, `AND`, `OR`.

The `AND` operator is only true if both arguments are true.

```python
print(True and True)
print(True and False)
print(False and True)
print(False and False)
```

The `OR` operator is true if either argument is true.

```python
print(True or True)
print(True or False)
print(False or True)
print(False or False)
```

The `NOT` operator is used to invert a logical value.

```python
print(not(True))
print(not(False))
```
