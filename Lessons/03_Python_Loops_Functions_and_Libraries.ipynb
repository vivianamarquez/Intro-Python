````python
# Today we will learn how to do things like this with our own data structures.

import random

def guess_capital_game(countries_capitals):
    # Select a random country from the dictionary.
    country = random.choice(list(countries_capitals.keys()))

    # Ask the user for the capital of the selected country.
    user_answer = input(f"What is the capital of {country}? ")

    # Check whether the user's answer is correct.
    accepted_answers = [capital.strip().lower() for capital in countries_capitals[country].split(",")]
    if user_answer.strip().lower() in accepted_answers:
        print(f"✅✅✅ Correct! Well done.\nThe capital of {country} is {countries_capitals[country]}.")
    elif user_answer.strip().lower() == "i don't know" or user_answer.strip().lower() == "":
        print(f"🙆‍♀️ No problem. We learn together.\nThe capital of {country} is {countries_capitals[country]}.")
    else:
        print(f"❌❌❌ Incorrect.\nThe capital of {country} is {countries_capitals[country]}.")

# Example use
guess_capital_game(countries_capitals)
````

# ⚙️ Functions

Functions are pieces of code that you can use again and again instead of writing them multiple times. They make your life easier!


Ingredients of a function:
- Input data
- A name
- Instructions
- Result

A function has four main ingredients:

```python
def function_name(input_data):
    # instructions
    return result
```

- The function name tells Python which reusable block to run.
- The input data goes inside the parentheses.
- The indented code block contains the instructions.
- `return` sends the result back.

To use a function that already exists, call its name and pass values inside the parentheses:

```python
function_name("hello", 3141592, [23, 7])
```

````python
def say_hello(name):
  print(f"Hello {name}!")
  print("How are you?")
````

````python
say_hello("Victoria")
````

````python
def add_five(number):
  return number + 5
````
