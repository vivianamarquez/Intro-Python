## Exercise #1

The school day is seven and a half hours long, including one hour for lunch. You are taking 5 classes. Write an equation to find out how many hours you spend in each class.

• Think about it: what type of division will you want to use? Why?

````python
school_day_hours = 7.5
lunch_hour = 1
number_of_classes = 5
````

````python
class_hours = (school_day_hours - lunch_hour) / number_of_classes
````

````python
print(f"I spend {class_hours} hours in each class")
````

**Other functions related to text strings**

- `str()` - Converts an integer/float into a text string.
- `int()` - Converts a text string into an integer.
- `float()` - Converts a text string into a floating-point number.
- `len()` - Tells you how many characters are in a text string.
- `.upper()` - Converts all letters in a text string to UPPERCASE.
- `.lower()` - Converts all letters in a text string to lowercase.
- `.swapcase()` - Converts uppercase letters to lowercase and lowercase letters to uppercase.
- `.capitalize()` - Converts the first letter of a text string to uppercase.
- `.strip()` - Removes any whitespace from the beginning or end of a text string.
- `.split(",")` - Splits a text string into substrings.
- `.startswith()` - Returns true if the text string starts with the specified value.
- `.endswith()` - Returns true if the text string ends with the specified value.
- `.index()` - Finds the first occurrence of the specified value.
- `.isnumeric()` - Checks whether the text string contains numeric characters.
- `==` - Checks whether 2 text strings are equal.
- `!=` - Checks whether 2 text strings are NOT equal.
- `in` / `not in` - Checks whether a substring is or is not inside the text string.

## Tuples

A tuple is like a list, but with two differences:

- It is not mutable, so it cannot be changed.
  - You cannot add new elements.
  - You cannot delete elements.
  - You cannot modify elements.
- It is declared with parentheses `(` `)`.

````python
coordinates = ('2.44 N', '76.61 W')
````

````python
coordinates[0]
````

````python
coordinates[0] = '3.44 N'
````

**Advantages**
- Tuples are faster than lists because they are not dynamic.
- They make your code safer for data that does not need to be modified.

## `For`

Used to repeat code a **specific number** of times.

```python
for item in iterable_variable:
  # do something with that item
```

- To iterate means to repeat a process, or a set of instructions, several times.
- The iterable variable can be a list, a text string, or a range.

````python
odds = [1, 3, 5, 7, 9, 11, 13]

for odd in odds:
  print(odd)
````

````python
odds = [1, 3, 5, 7, 9, 11, 13]

for odd in odds:
  print(odd + 1)
````

````python
odds = [1, 3, 5, 7, 9, 11, 13]

for sandwich in odds:
  print(sandwich + 1)
````

````python
list(range(1, 21))
````

````python
for box in range(11):
    print(f"Organized box number {box + 1}")
````

### `While`

Used to repeat code **until** a condition is met.

```python
while (boolean):
    # run this code
```

````python
clean_room = False
while (clean_room != True):
    print("Pick up your socks!")
````

### ⚠️ Be Careful with Infinite Loops!

````python
counter = 5

while (counter >= 0):
    print(counter)
    counter = counter - 1

print('Done!')
````

# 🚀 List Comprehension

```python
[operation for item in list]
```

- List comprehension in Python is a quick, compact way to generate a list from a string or any other list.
- It can replace loops that modify lists.
- It applies the expression to each element in the list.

````python
students = ['Jessy', 'Kathy', 'Sammy']
````

````python
for student in students:
    print(student.upper())
````

````python
[student.upper() for student in students]
````
